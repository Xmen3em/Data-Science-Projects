```python
str_col = df.select_dtypes('object').columns.to_list()
num_col = df.columns.to_list()
for col in str_col:
  num_col.remove(col)
print(num_col)
```

```python
num_data = df[num_col]
correlation = num_data.corr()
correlation['TotalPayBenefits'].sort_values(ascending = False)
```
